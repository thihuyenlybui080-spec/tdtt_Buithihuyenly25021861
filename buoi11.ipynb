```python
#3
def sum_square_digit(n):
  tong=0
  while n>0:
    so = n % 10
    tong = so * so
    n //= 10
  return tong
def is_happy(a):
  seen = set()
  while a != 1:
    if a in seen :
      return False
    seen.add(a)
    a = sum_square_digit(a)
  return True
try:
  a=int(input())
  if a <= 0:
    print("Nhập sai")
  else:
    if is_happy(a):
      print("Yes")
    else:
      print("No")
except ValueError:
  print("ValueError: Dữ liệu nhập vào không hợp lệ")
```
