```python
kwh = int(input(' nhap vao so kwh dien tieu thu:'))
tien = 0

if kwh <= 50:
    tien = kwh * 1500

elif kwh <= 100:
    tien = 50 * 1500 + (kwh - 50) * 2000

else:
    tien = 50 * 1500 + 50 * 2000 + (kwh - 100) * 3000

print("Tien dien:", tien, "đ")
```
