```python
#6
from math import isqrt
def is_primes(a):
    if a < 2:
        return False
    for x in range(2, isqrt(a) + 1):
        if a % x == 0:
            return False
    return True
def sum_primes(a, b):
    tong = 0
    for n in range(a, b + 1):
        if is_primes(n):
            tong += n
    return tong
a, b = map(int, input().split())
results = sum_primes(a, b)
print(results)
```

```python
#7
from math import isqrt
def is_prime(a):
    if a < 2:
        return False
    for i in range(2, isqrt(a) + 1):
        if a % i == 0:
            return False
    return True
def largest_prime_factor(n):
    s =[]
    for x in range (1, isqrt(n) + 1):
        if n % x == 0:
            if is_prime(x):
                s.append(x)
            other = n // x
            if is_prime(other):
                s.append(other)
    return s
numbers = int(input())
results = largest_prime_factor(numbers)
print(max(results))
```

```python
#8
def reverse(n):
    new = 0
    while n > 0:
        new = new * 10 + n % 10
        n //= 10
    return new
def palindrome(a):
    buoc = 0
    while a != reverse(a):
        buoc += 1
        a = a + reverse(a)
    return buoc, a
if __name__ == "__main__":
    numbers = int(input())
    buoc, a = palindrome(numbers)
    print(buoc, a, sep= ' ')
```

```python
#9
from math import isqrt
def perfect_square(n):
    tim_thay = False
    for x in range(1, n):
        if x == pow(isqrt(x), 2):
            tim_thay = True
            print(x, end= " ")
    if not tim_thay:
        return "No number"
    return ""
numbers = int(input())
results = perfect_square(numbers)
print(results)
```

```python
#10
def collatz(n):
    buoc = 1
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
        buoc += 1
    return buoc
def long(i):
    L = -1
    min_num = 1
    for x in range(1, i + 1):
        dem = collatz(x)
        if L < dem or ( L == dem and x < min_num):
            L = dem
            min_num = x
    return min_num, L

numbers = int(input())
min_num, L= long(numbers)
print(min_num, L)
```

```python
#11
from math import isqrt
def even_factor(n):
    dem = 0
    if 0 < n < 10 ** 6:
        for x in range(1, isqrt(n) + 1):
            if n % x == 0:
                other = n // x
                if x % 2 == 0:
                    dem += 1
                if other != x and other % 2 == 0:
                    dem += 1

        return dem
numbers = int(input())
results = even_factor(numbers)
print(results)
```

```python
#13
from math import isqrt
def close_pair_number(n):
    tong = 0
    for x in range(1, isqrt(n) + 1 ):
        if n % x == 0:
            other = n // x
            tong += x
            if other != x and other != n:
                tong += other
    return tong
def close(a, b):
    if close_pair_number(a) == b and close_pair_number(b) == a:
        return True
    else:
        return False
a, b = map(int, input().split())
results = close(a, b)
print(results)
```

```python
#14
from math import isqrt
def max_factor(n):
    s = set()
    for x in range(1, isqrt(n) + 1):
        if n % x == 0:
            s.add(x)
            other = n // x
            s.add(other)
    return s
def same_fac(a, b):
    return max_factor(a) & max_factor(b)
a, b = map(int, input().split())
results = same_fac(a, b)
print(max(results))
```

```python
#15
def chicken_and_dog(a, b):
    for ga in range(1, a + 1):
        cho = a - ga
        if cho * 4 + ga * 2 == b:
            return f"{ga} {cho}"
    return "invalid"
a, b = map(int, input().split())
results = chicken_and_dog(a, b)
print(results)
```
