```python
#1
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (right + left) // 2
        if arr[mid] < target:
            left = mid + 1
        elif arr[mid] > target:
            right = mid - 1
        else:
            return mid
    return -1
arr = list(map(int, input().strip().split()))
target = int(input())
print(binary_search(arr, 0, len(arr) -1, target))
```

```python
#2
def count_occurrences(arr, x):
    dem = 0
    for i in range(len(arr)):
        if arr[i] == x:
            dem += 1
    return dem
arr = list(map(int, input().strip().split()))
x = int(input())
print(count_occurrences(arr, x))
```

```python
#3
def bubble_sort(n):
    for i in range(len(n)):
        swapped = False
        for j in range(0, len(n) - i - 1):
            if n[j] > n[j + 1]:
                n[j], n[j + 1] = n[j + 1], n[j]
                swapped = True
        if not swapped:
            break
    return n
n = list(map(int, input().strip().split()))
print(*bubble_sort(n))
```

```python
#3
def selection_sort(n):
    for i in range(len(n)):
        min_idx = i
        for j in range(i + 1, len(n)):
            if n[j] < n[min_idx]:
                min_idx = j
        n[i], n[min_idx] = n[min_idx], n[i]
    return n
n = list(map(int, input().strip().split()))
print(*selection_sort(n))
```

```python
#4
def xhien(a):
    a.sort()
    cnt = 1
    ans = 1
    res = a[0]
    for x in range(1, len(a)):
        if a[x] == a[x - 1]:
            cnt += 1
        else:
            if cnt > ans:
                ans = cnt
                res = a[x - 1]
    return f"{ res } xuat hien nhieu nhat, som nhat, {cnt} lan"
a = list(map(int, input().strip().split()))
print(xhien(a))
```

```python
#5
def count_pairs(arr, x):
  count = 0
  for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
      if arr[i] + arr[j] == x:
        count += 1
  return count
s = list(map(int, input().split()))
x = int(input())
print(count_pairs(s, x))
```

```python
#7
def lengthOfLIS(nums):
    n = len(nums)
    if n == 0:
        return 0

    dp = [1] * n

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
```

```python
#13
def harmonic_sequence(a):
    a.sort()
    b=list(set(a))
    b.sort()
    max_len=0
    for i in range(len(b)-1):
        if b[i+1] - b[i] == 1:
            leng = a.count(b[i]) + a.count(b[i]+1)
            if leng > max_len:
                max_len = leng
    return max_len
a=list(map(int,input().split()))
print(harmonic_sequence(a))
```
