```python
#1
def trung_nhau(n):
    kq = []
    for x in n:
        if x not in kq:
            kq.append(x)
    return kq
n = list(map(int, input().strip().split()))
print(trung_nhau(n))
```

```python
#3
def xoay_trai(n, k):
    return n[k:] + n[:k]
n = list(map(int, input().strip().split()))
k = int(input())
print(tuple(xoay_trai(n, k)))
```

```python
#4
a = input().split()
results = {}
for pair in a:
    key, Value = pair.split(":")
    if key not in results:
        results[key] = []
    results[key].append(Value)
print(results)
```

```python
#5
a = list(map(int, input().strip().split()))
results = {'positives': 0,
            'negatives': 0,
            'zeros': 0}
for x in a:
    if x < 0:
        results['negatives'] += 1
    elif x == 0:
        results['zeros'] += 1
    else:
        results['positives'] += 1
print(results)
```

```python
#9
d1 = eval(input())
d2 = eval(input())
result = {}
for key in d1:
    result[key] = d1[key]
for key in d2:
    if key in result:
        result[key] += d2[key]
    else:
        result[key] = d2[key]
sort_result = dict(sorted(result.items()))
print(sort_result)
```

```python
#10
s = list(map(int, input().strip().split()))
k = int(input())
tong = []
n = len(s)
for x in range(n):
    for i in range(x + 1, n):
        if s[x] + s[i] == k:
            tong.append((s[x], s[i]))
def lay_ptu_dau(x):
    return x[0]
tong.sort(key = lay_ptu_dau)
print(tong)
```

```python
#12
def xuat_hien(a):
    min = a[0]
    lan = 0
    for x in a:
        tong = 0
        for y in a:
           if y == x:
                tong += 1
        if tong > lan or (tong == lan and min > x):
            lan = tong
            min = x
    return min
a = list(map(int, input().strip().split()))
print(xuat_hien(a))
```

```python
#13
def reverse_keyvalue(c):
    a = c[1::2]
    b = c[0::2]
    return dict(zip(a, b))
c = input().split()
print(reverse_keyvalue(list(c)))
```

```python
#20
def ptu_dtien(a, k):
    for x in range(len(a)):
        if a[x] == k:
            return x + 1
    return -1
a = list(map(int, input().strip().split()))
k = int(input())
print(ptu_dtien(a, k))
```
