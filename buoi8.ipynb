```python
#1
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
        self.pi = 3.14

    def dien_tich_toan_phan(self):
        return 2 * self.pi * self.r * (self.r + self.h)

    def the_tich(self):
        return self.pi * self.r * self.r * self.h

r, h = map(float, input().split())
c = Cylinder(r, h)
print(f"{c.dien_tich_toan_phan():.2f} {c.the_tich():.2f}")
```

```python
#2
class Date:
    def __init__(self, d, m, y):
        self.day = d
        self.month = m
        self.year = y
    def nam_nhuan(self):
        return self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0)

    def days_in_month(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        if self.month in [4, 6, 9, 11]:
            return 30
        if self.month == 2:
            return 29 if self.nam_nhuan() else 28
        return 0

    def is_valid(self):
        if self.year < 1 or self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > self.days_in_month():
            return False
        return True

    def next_date(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1
d, m, y = map(int, input().split())
date = Date(d, m, y)
if date.is_valid():
    date.next_date()
    print(f"{date.day}/{date.month}/{date.year}")
```

```python
#3
from math import sqrt
class point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def check(self):
    if self.x == 0 and self.y == 0:
      print("là gốc tọa độ")
    elif self.x == 0 and self.y != 0:
      print("là điểm nằm trên trục tung")
    elif self.x != 0 and self.y == 0:
      print("là điểm nằm trên trục hoành")
    else:
            print("khong nam tren truc nao")
  def kc(self):
    kc = sqrt(self.x**2 + self.y**2)
    return kc
x, y = map(int, input().split())
p = point(x, y)
p.check()
print(f"{p.kc():.2f}",)
```

```python
#4
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def set_numbers(self, a, b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
    def product(self):
        return self.a * self.b
    def divide(self):
        if self.b == 0:
            return "Khong the chia cho 0"
        return self.a / self.b
    def power(self):
        return self.a ** self.b
    def mod(self):
        if self.b == 0:
            return "Khong the chia cho 0"
        return self.a % self.b
a, b = map(float, input("Nhap hai so: ").split())
cal = Calculator(a, b)
while True:
    print("1. Cong")
    print("2. Tru")
    print("3. Nhan")
    print("4. Chia")
    print("5. Luy thua")
    print("6. Chia lay du")
    print("7. Cap nhat 2 so moi")
    choice = input("Chon phep toan (1-7): ")
    if choice == "1":
        print("Ket qua:", cal.add())
    elif choice == "2":
        print("Ket qua:", cal.subtract())
    elif choice == "3":
        print("Ket qua:", cal.product())
    elif choice == "4":
        print("Ket qua:", cal.divide())
    elif choice == "5":
        print("Ket qua:", cal.power())
    elif choice == "6":
        print("Ket qua:", cal.mod())
    elif choice == "7":
        a, b = map(float, input("Nhap lai hai so: ").split())
        cal.set_numbers(a, b)
    else:
        print("Lua chon khong hop le")
    stop = input("Ban co muon thoat? ")
    if stop == "yes" or stop == "YES" or stop == "Yes":
        break
```

```python
#5
class ShoppingCart:
    def __init__(self):
        self.products = []
    def add_product(self, name, price):
        product = {"name": name, "price": price}
        self.products.append(product)
    def remove_product(self, name):
        for p in self.products:
            if p["name"] == name:
                self.products.remove(p)
                return True
        return False
    def is_empty(self):
        return len(self.products) == 0
    def total_price(self):
        total = 0
        for p in self.products:
            total += p["price"]
        return total
    def show_products(self):
        if self.is_empty():
            print("Gio hang rong")
        else:
            for p in self.products:
                print(p["name"], "-", p["price"])
    def clear_cart(self):
        self.products.clear()
cart = ShoppingCart()
while True:
    print("2. Them san pham")
    print("3. Xoa san pham theo ten")
    print("4. Kiem tra gio hang rong")
    print("5. Tinh tong tien")
    print("6. Hien thi gio hang")
    print("7. Xoa toan bo gio hang")
    choice = input("Nhap lua chon (2-7): ")
    if choice == "2":
        name = input("Nhap ten san pham: ")
        price = float(input("Nhap gia san pham: "))
        cart.add_product(name, price)
    elif choice == "3":
        name = input("Nhap ten san pham can xoa: ")
        if cart.remove_product(name):
            print("Da xoa san pham")
        else:
            print("Khong tim thay san pham")
    elif choice == "4":
        if cart.is_empty():
            print("Gio hang dang rong")
        else:
            print("Gio hang khong rong")
    elif choice == "5":
        print("Tong tien:", cart.total_price())
    elif choice == "6":
        cart.show_products()
    elif choice == "7":
        cart.clear_cart()
        print("Da xoa toan bo gio hang")
    else:
        print("Lua chon khong hop le")
    stop = input("Co muon thoat khoi chuong trinh khong? ")
    if stop == "yes":
        break
```

```python
#6
class Rectangle:
  def __init__(self, w, h):
    self.w = w
    self.h = h
  def area(self):
    return self.w * self.h
  def perimeter(self):
    return 2 * (self.w + self.h)
  def scale(self, k):
    self.w *= k
    self.h *= k
w, h, k = map(float, input().split())
rect = Rectangle(w, h)
rect.scale(k)
print(f"{rect.area():.2f} {rect.perimeter():.2f}")
```

```python
#7
from math import*
class Fraction:
  def __init__(self,tu_1,mau_1,oper,tu_2,mau_2):
    self.tu_1=tu_1
    self.tu_2=tu_2
    if mau_1 == 0 or mau_2 == 0:
      raise ValueError('Mẫu khác 0')
    self.mau_2=mau_2
    self.mau_1=mau_1
    self.oper = oper
  def add(self):
    if self.mau_1 == self.mau_2:
      return f'{(self.tu_1+self.tu_2)//gcd(self.tu_1+self.tu_2,self.mau_1)}/{self.mau_1}'
    else:
      bcnn = lcm(self.mau_1,self.mau_2)
      self.tu_1 = bcnn // self.mau_1
      self.tu_2 = bcnn // self.mau_2
      return f'{(self.tu_1 + self.tu_2)//gcd(self.tu_1 + self.tu_2,bcnn)}/{bcnn // gcd(self.tu_1 + self.tu_2,bcnn)}'
  def sub(self):
    if self.mau_1 == self.mau_2:
      return f'{(self.tu_1-self.tu_2)//gcd(self.tu_1-self.tu_2,self.mau_1)}/{self.mau_1}'
    else:
      bcnn = lcm(self.mau_1,self.mau_2)
      self.tu_1 = bcnn // self.mau_1
      self.tu_2 = bcnn // self.mau_2
      return f'{(self.tu_1 - self.tu_2)//gcd(self.tu_1 - self.tu_2,bcnn)}/{bcnn // gcd(self.tu_1 - self.tu_2,bcnn)}'
  def mul(self):
    return f'{(self.tu_1 * self.tu_2)//gcd(self.tu_1 * self.tu_2,self.mau_1 * self.mau_2)}/{(self.mau_1 * self.mau_2)//gcd(self.tu_1 * self.tu_2,self.mau_1 * self.mau_2)}'
  def div(self):
    return f'{(self.tu_1 * self.mau_2)//gcd(self.tu_1 * self.mau_2,self.tu_2 * self.mau_1)}/{(self.mau_1 * self.tu_2)//gcd(self.tu_1 * self.mau_2,self.tu_2 * self.mau_1)}'
  def menu(self):
    if self.oper == '+':
      print(self.add())
    elif self.oper == '-':
      print(self.sub())
    elif self.oper == '*':
      print(self.mul())
    elif self.oper == '/':
      print(self.div())
tu_1,mau_1,oper,tu_2,mau_2 = map(str,input().split())
s=Fraction(int(tu_1),int(mau_1),oper,int(tu_2),int(mau_2))
s.menu()
```

```python
#8
class BankAccount:
  def __init__(self,owner,balance):
    self.owner=owner
    self.balance=balance
  def deposit(self,dep):
    self.balance += dep
  def withdraw(self,wdr):
    if self.balance >= wdr:
      self.balance -= wdr
  def __str__(self):
    return ('%.2f' %self.balance)
owner,balance = map(str,input().split())
s=BankAccount(owner,float(balance))
n=int(input())
for _ in range(n):
  method, money = map(str,input().split())
  money = float(money)
  if method == 'WITHDRAW':
    s.withdraw(money)
  elif method == 'DEPOSIT':
    s.deposit(money)
print(s)
```

```python
#9
class Student:
  def __init__(self,name):
    self.name = name
    self.scores = []
  def add(self,sub,score):
    self.scores.append((sub,score))
  def avg(self):
    b=[x[1] for x in self.scores]
    return round(sum(b)/len(b),2)
  def rank(self):
    if self.avg() >=8:
      return 'Excellent'
    elif 6.5<=self.avg() < 8:
      return 'Good'
    elif 5 <= self.avg() < 6.5:
      return 'Average'
    else:
      return 'Poor'
name=input()
s=Student(name)
n=int(input())
for _ in range(n):
  sub,score = map(str,input().split())
  s.add(sub,float(score))
print(f'{name} {s.avg():.2f} {s.rank()}')
```

```python
#11
from math import*
class space:
  def __init__(self,a1,a2,a3,b1,b2,b3,c1,c2,c3,d1,d2,d3):
    self.a1=a1
    self.a2=a2
    self.a3=a3
    self.b1=b1
    self.b2=b2
    self.b3=b3
    self.c1=c1
    self.c2=c2
    self.c3=c3
    self.d1=d1
    self.d2=d2
    self.d3=d3
  def mp1(self):
    n1,n2,n3=self.a1-self.b1,self.a2-self.b2,self.a3-self.b3
    k1,k2,k3=self.a1-self.c1,self.a2-self.c2,self.a3-self.c3
    return [n2*k3-k2*n3,n3*k1-k3*n1,n1*k2-k1*n2]
  def mp2(self):
    n1,n2,n3=self.c1-self.b1,self.c2-self.b2,self.c3-self.b3
    k1,k2,k3=self.d1-self.c1,self.d2-self.c2,self.d3-self.c3
    return [n2*k3-k2*n3,n3*k1-k3*n1,n1*k2-k1*n2]
  def goc(self):
    a,b,c=self.mp1()
    x,y,z=self.mp2()
    tu = abs(a*x + b*y + c*z)
    mau= sqrt(a**2 + b**2 +c**2) * sqrt(x**2 +y**2 +z**2)
    return round(degrees(acos(tu/mau)),2)
a1,a2,a3 = map(float,input().split())
b1,b2,b3 = map(float,input().split())
c1,c2,c3 = map(float,input().split())
d1,d2,d3 = map(float,input().split())
s = space(a1,a2,a3,b1,b2,b3,c1,c2,c3,d1,d2,d3)
print(s.goc())
```

```python
#12
class ComplexNumber:
  def __init__(self,real1,imag1,real2,imag2):
    self.real1 = real1
    self.imag1 = imag1
    self.real2 = real2
    self.imag2 = imag2
  def mul(self):
    return f'{self.real1*self.real2 - self.imag1*self.imag2} {self.real1*self.imag2 + self.imag1*self.real2}'
real1,imag1 = map(int,input().split())
real2,imag2 = map(int,input().split())
s = ComplexNumber(real1,imag1,real2,imag2)
print(s.mul())
```

```python
#13
from math import*
class Triangle:
  def __init__(self,edge1,edge2,edge3):
    self.edge1=edge1
    self.edge2=edge2
    self.edge3=edge3
  def check(self):
    if (self.edge1 <= 0 or self.edge2 <= 0 or self.edge3 <= 0) or (self.edge1 + self.edge2 <= self.edge3 or self.edge1 + self.edge3 <= self.edge2 or self.edge3 + self.edge2 <= self.edge1):
      return False
    else: return True
  def dientich(self):
    if self.check():
      p = (self.edge1 + self.edge2 + self.edge3)/2
      s= sqrt(p*(p-self.edge1)*(p-self.edge2)*(p-self.edge3))
      return f'{s:.2f}'
    else:
      return "invalid"
a,b,c=map(float,input().split())
s = Triangle(a,b,c)
print(s.dientich())
```
