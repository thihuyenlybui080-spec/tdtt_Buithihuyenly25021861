```python
#1
def max_of_two(a, b):
    return max(a, b)
a,b = map(int, input().split())
print(max_of_two(a, b))
```

```python
#2
def swap(a, b):
    a, b = b, a
    return a, b
a, b = map(int, input().split())
print(swap(a, b))
```

```python
#3
from math import isqrt
def so_ngto(n):
    if n < 2:
        return False
    for x in range(2, isqrt(n) + 1):
        if n % x == 0:
            return False
    return True
n = int(input())
print(so_ngto(n))
```

```python
#4
from math import isqrt
def hoan_hao(n):
    tong = 0
    for x in range(1, isqrt(n) + 1):
        if n % x == 0:
            tong += x
        other = n // x
        if other != x and other != n:
            tong += other
    if tong == n:
        return True
    return False
n = int(input())
print(hoan_hao(n))
```

```python
#5
def list_so(n, k):
    for x in range(len(n)):
        if n[x] == k:
            return x + 1
    return -1
n = list(map(int, input().strip().split()))
k = int(input())
print(list_so(n, k))
```

```python
#6
def giai_thua(n):
    tich = 1
    for x in range(1, n + 1):
        tich *= x
    return tich
n = int(input())
print(giai_thua(n))
```

```python
#7
def caculator(num1, num2, operat):
    if operat == "+":
        result = num1 + num2
    elif operat == "-":
        result = num1 - num2
    elif operat == "*":
        result = num1 * num2
    elif operat == "/":
        if num2 == 0:
            return "khong hop le"
        result = num1 / num2
    else:
        return "ko hop le"
    return result
s = input().split()
num1 = float(s[0])
num2 = float(s[2])
operat = s[1]
print(round(caculator(num1, num2, operat), 2))
```

```python
#8
def hamming_distance(x, y):
  return bin(x ^ y).count("1")
x, y = map(int, input().split())
print(hamming_distance(x, y))
```

```python
#9
def tong_chu_so(n):
  tong = 0
  while n > 0:
    tong += n % 10
    n //= 10
  return tong

n = int(input())
print(tong_chu_so(n))
```

```python
#10
def dang_cau(a, b):
    mapping = {}
    used_values = set()
    for x, y in zip(a, b):
        if x in mapping:
            if mapping[x] != y:
                return False
        else:
            if y in used_values:
                return False
            mapping[x] = y
            used_values.add(y)
    return True
a, b = input().split()
print(dang_cau(a, b))
```
